```python
import pandas as pd
import sqlite3
import requests
from io import StringIO

DB_NAME="banco.db"


def corre_sql( sql_text):
  conn = sqlite3.connect(DB_NAME)
  df_select = pd.read_sql_query(sql_text, conn)
  return df_select
```
